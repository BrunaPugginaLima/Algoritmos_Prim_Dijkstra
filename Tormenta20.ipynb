```python
from google.colab import drive
drive.mount('/content/drive')

!pip install -q pymupdf
import fitz  # PyMuPDF
import google.generativeai as genai
from google.colab import userdata

# Configurar a API Gemini
GOOGLE_GEMINI_API_KEY = userdata.get('GOOGLE_GEMINI_API_KEY')
genai.configure(api_key=GOOGLE_GEMINI_API_KEY)

# Caminho do PDF
caminho_pdf = "/content/drive/MyDrive/Colab Notebooks/Tormenta 20 - Livro Base Completo - Biblioteca do Duque.pdf"

# Extrair texto do PDF
texto = ""
try:
    with fitz.open(caminho_pdf) as doc:
        for pagina in doc:
            texto += pagina.get_text()
except Exception as e:
    print(f"Erro ao extrair texto do PDF: {e}")
    exit()

print(f"Tamanho do texto extraído: {len(texto)} caracteres")

# Dividir o texto em partes menores (100.000 caracteres por parte)
tamanho_parte = 100000
partes = [texto[i:i + tamanho_parte] for i in range(0, len(texto), tamanho_parte)]
print(f"Texto dividido em {len(partes)} partes")

# Inicializar o modelo Gemini
modelo = genai.GenerativeModel("gemini-1.5-flash-latest")

# Função para encontrar a parte relevante com base na pergunta
def encontrar_parte_relevante(pergunta, partes):
    # Palavras-chave relacionadas à pergunta (você pode expandir esta lista)
    palavras_chave = {
        "classes": ["classe", "personagem", "guerreiro", "mago", "ladino"],
        "magias": ["magia", "feitiço", "arcana", "divina"],
        "combate": ["combate", "ataque", "defesa", "dano"],
        # Adicione mais tópicos conforme necessário
    }

    # Identificar palavras-chave na pergunta
    for topico, palavras in palavras_chave.items():
        if any(palavra in pergunta.lower() for palavra in palavras):
            # Procurar a parte do texto que contém as palavras-chave
            for i, parte in enumerate(partes):
                if any(palavra in parte.lower() for palavra in palavras):
                    return parte, i
    # Se não encontrar palavras-chave específicas, retornar a primeira parte
    return partes[0], 0

    # Definir o contexto inicial com todo o texto (ou uma parte maior)
    conteudo_base = "".join(partes)[:30000]
```
